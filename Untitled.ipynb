# K-means

## Recency

```python
recency= df.groupby("CustomerID").InvoiceDate.max().reset_index()
recency["Recency"]=(recency["InvoiceDate"].max()-recency["InvoiceDate"]).dt.days
recency
```

```python
recency["Recency"]=(recency["InvoiceDate"].max()-recency["InvoiceDate"]).dt.days
```

```python
recency.head()
```

```python
del recency['InvoiceDate']
```

```python
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"max_iter": 300,
"random_state": 42,
}

sse = []
for k in range(1, 11):
 kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
 kmeans.fit(recency)
 sse.append(kmeans.inertia_)
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
label = kmeans.fit_predict(recency[['Recency']])
print(label)
```

```python
plt.scatter(recency.values[:, 0], recency.values[:, 1], c=label, s=50, cmap='viridis')
plt.show()
```

```python
overall = pd.DataFrame(df['CustomerID'].unique())
overall.columns = ['CustomerID']
```

```python
overall = pd.merge(overall, recency[['CustomerID','Recency']], on='CustomerID')
overall['RecenClus']=kmeans.fit_predict(overall[['Recency']])
overall.head()
```

```python
overall.groupby('RecenClus')['Recency'].describe()
```

```python
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final
```

```python
overall = order_cluster('RecenClus', 'Recency',overall,False)
overall.groupby('RecenClus')['Recency'].describe()
```

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(frequency[['Freq']])
```

```python
plt.scatter(frequency.values[:, 0], frequency.values[:, 1], c=label, s=50, cmap='viridis')
plt.show()
```

```python
overall = pd.merge(overall, frequency[['CustomerID','Freq']], on='CustomerID')
overall['FreqClus'] = kmeans.fit_predict(overall[['Freq']])
```

```python
overall = order_cluster('FreqClus', 'Freq',overall,True)
overall.groupby('FreqClus')['Freq'].describe()
```

# Hierarchical Clustering

```python
sub_df=overall[['CustomerID', 'Recency', 'Freq', 'Rev']]
sub_df
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(sub_df)
data_scaled = pd.DataFrame(data_scaled, columns=sub_df.columns)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
sub_df['Segment']=cluster.fit_predict(recency)
```

```python
sub_df.head()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Recency'], data_scaled['Freq'], c=cluster.labels_) 
plt.title('Recency and Frequency')
plt.xlabel('Recency')
plt.ylabel('Frequency')
```
